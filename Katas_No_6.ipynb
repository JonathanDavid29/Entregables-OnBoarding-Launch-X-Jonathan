```python
planetas=['Mercury','Venus','Tierra','Marte','Jupiter','Saturno','Neptuno']
Planeta_usuario= input('Ingresa el nombre del planeta :) \nPrimer letra en mayusculas')
planet_index = planetas.index(Planeta_usuario)

print('Los planetas mas cercanos al sol con respecto a ' + Planeta_usuario)
print(planetas[0:planet_index])

print('Los planetas mas lejanos al sol con respecto a ' + Planeta_usuario)
print(planetas[planet_index + 1:])
```
